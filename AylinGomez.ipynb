```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/aylinggq/BCIS-321/refs/heads/main/top_100_books-2020-Aylin_Gomez_Quezada.csv')
print(df)
```

```python
#Remove columns that are blank
df.drop(["Unnamed: 4"], axis=1, inplace=True)
df.drop(["Unnamed: 0"], axis=1, inplace=True)
df.drop(["Unnamed: 9"], axis=1, inplace=True)
df.drop(["Unnamed: 10"], axis=1, inplace=True)
print(df)
#Using a loop update the price of the book titled “Man’s Search for Meaning” to $9.99
df.loc[86]=[88, 'Man\'s Search for Meaning','Vicktor Frankl', 4.7, 4959, 'Paperback', 9.99]
df.head(10)
```

```python
#Using a loop make sure the type field is either Hardcover or Paperback (convert Board Book to Hardcover and Mass Market Paperback to Paperback)
changed = 0
i = 0
for t in df.type:
  if t == 'Board Book':
    df.type.replace('Board Book', 'Hardcover', inplace=True)
    changed +=1
  if t == 'Mass Market Paperback':
    df.type.replace('Mass Market Paperback', 'Paperback', inplace=True)
    changed +=1
print(changed)
df.head()
```

```python
#Using a loop create a new calculated field representing rating/price (that is the rating divided by the price rounded to two decimal
#places e.g. 4.2 / 15.57 = 0.27). Create this field, call it “rating_by_price”, and add it to the DataFrame as the fifth column (on the right-hand side of rating)
rating = df['rating']
price = df['price']
rating_by_price = rating/price
print(rating_by_price)
df.insert(4, 'rating_by_price', rating_by_price)
df.sort_values(by=['rating_by_price'], ascending=False)
df.head()
```

```python
#Write your updated DataFrame to a new CSV file called “{Your_Name}_top_100_books_cleaned.csv”
df.to_csv('Aylin_Gomez_top_100_books_2020_cleaned.csv')
```
